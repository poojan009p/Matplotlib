```python
import seaborn as sns
import matplotlib.pyplot as plt  
```

### we use distribution plot for quantitative data (continous data)(Age,salary,etc)

###  plots use to visualize quantitative data 
 * histplot
 * jointplot
 * pairplot
 * rugplot

```python
df = sns.load_dataset("tips")
```

```python
df.shape
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"],kde=True)

```

###  Two compare two data columns

```python
sns.jointplot(x = 'total_bill', y='tip',data = df,kind="scatter")
```

```python
sns.jointplot(x = 'total_bill', y='tip',data = df,kind="kde")
```

```python
sns.jointplot(x = 'total_bill', y='tip',data = df,kind="hex")
```

```python
sns.jointplot(x = 'total_bill', y='tip',data = df,kind="reg")
```

## pair plot 

```python
sns.pairplot(df,hue="sex")
```

```python
sns.rugplot(df["total_bill"])
```
