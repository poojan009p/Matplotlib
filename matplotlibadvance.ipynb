```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore")
```

## Matplotlib basics

```python
x = np.linspace(0,10,9)
x
```

```python
y = x **2
y
```

### linegraph

```python
plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

## Subplots in Matplotlibs 

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("HELLO")
plt.subplot(2,2,4)
plt.title("Poojan")
plt.plot(x**2,y**3)
plt.subplot(2,2,2)
plt.plot(x**2,y)
plt.subplot(2,2,3)
plt.plot(y,x*4)
```

## Object oriented way 

```python
fig =plt.figure(figsize=(6,8))
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(y,x)
axis1=fig.add_axes([1.1,0.1,.5,.5])
axis1.plot(x,y)
```
